```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 1, figsize=(100,15))
axes.set_xlim(0, 600)
axes.set_ylim(-75, 75)
axes.set_aspect(1)
axes.grid(True)

# The target cone width, expressed as atan of angle on either side of the search
# direction.
cone_atan = 0.1
# Distance for jump.
jump_distance = 500
# Number of backoff attempts
num_attempts = 15


# We draw the cone we want to search in.
axes.add_line(plt.Line2D([0, jump_distance], [0, jump_distance * cone_atan], c="red"))
axes.add_line(plt.Line2D([0, jump_distance], [0, -jump_distance * cone_atan], c="red"))

# We draw the circles representing where we actually search.
distance = jump_distance
last_distance = distance
for i in range(num_attempts):
  radius = distance * cone_atan
  axes.add_patch(plt.Circle((distance,0), radius))
  last_distance = distance
  distance -= radius * (1.4 * 1.013**i)
print(f"Last attempted distance (i.e. the distance below which we can't jump if we press the {jump_distance}km button): {last_distance}")
```
